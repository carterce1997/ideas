```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 2000

dt = .01

alpha = 2/3
beta = 4/3
gamma = 1
delta = 1

xt = .9
yt = .9
tt = 0

x = []
y = []
t = []

for i in range(n):
    
    x.append(xt)
    y.append(yt)
    t.append(tt)
    
    xt_next = xt + dt * (alpha * xt - beta * xt * yt)
    yt_next = yt + dt * (gamma * xt * yt - delta * yt)
    tt_next = tt + dt
    
    xt = xt_next
    yt = yt_next
    tt = tt_next
    
x = np.array(x)
y = np.array(y)
t = np.array(t)

fx = x / (x + y)
fy = y / (x + y)

entropy = -fx * np.log(fx) -fy * np.log(fy)

fig, ax = plt.subplots(2, 1)

ax[0].plot(t, x)
ax[0].plot(t, y)

ax[1].plot(t, entropy)

plt.show()
```
