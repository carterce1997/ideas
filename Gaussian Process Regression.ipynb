```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import gaussian_process
```

```python
x = np.linspace(0, 10, 100)
y_gt = 3 * x * np.sin(x)
y = y_gt + (5 + 2 * x) * np.random.standard_t(3, size=x.size)
```

```python
plt.scatter(x, y)
plt.plot(x, y_gt, color='red')
plt.show()
```

```python
gp = gaussian_process.GaussianProcessRegressor(alpha=2)
```

```python
gp.fit(x.reshape(-1,1), y)
```

```python
x_pred = np.linspace(-1, 11, 60)
y_pred = gp.predict(x_pred.reshape(-1, 1))
```

```python
plt.scatter(x, y)
plt.plot(x, y_gt, color='red')
plt.plot(x_pred, y_pred, color='orange')
plt.show()
```
